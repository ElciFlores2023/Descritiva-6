### EspaÃ§amento em torno das subplotagens

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='b', alpha=0.9)
        
# O mÃ©todo subplots_adjust ajusta o espaÃ§amento entre os eixos.
plt.subplots_adjust(wspace=0, hspace=0)
```

# Tiques, rÃ³tulos e legendas

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['ano 1', 'ano 2', 'ano 3', 'ano 4', 'ano 5'],
                            rotation=30, fontsize='small')
```

```python
ax.set_title('TÃ­tulo do grÃ¡fico')
ax.set_xlabel('PerÃ­odo')
```

#### adicionando legendas

```python
plt.close('all')
from numpy.random import randn
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
```

```python
ax.legend(loc='best')
```
